```python
# Load Dataset
order_items    = pd.read_csv("olist_order_items_dataset.csv")
orders         = pd.read_csv("olist_orders_dataset.csv")
order_payments = pd.read_csv("olist_order_payments_dataset.csv")
order_reviews  = pd.read_csv("olist_order_reviews_dataset.csv")
products       = pd.read_csv("olist_products_dataset.csv")
customers      = pd.read_csv("olist_customers_dataset.csv")
sellers        = pd.read_csv("olist_sellers_dataset.csv")
geolocation    = pd.read_csv("olist_geolocation_dataset.csv")
product_category_translation = pd.read_csv("product_category_name_translation.csv")
```

```python
order_items_check = []

for col in order_items.columns:
    order_items_check.append([col, order_items[col].dtype, order_items[col].isna().sum(), round((order_items[col].isna().sum()/len(order_items[col]))*100, 2),
                    order_items[col].duplicated().sum(),order_items[col].nunique(), list(order_items[col].drop_duplicates().sample(2).values)]);

order_items_desc = pd.DataFrame(columns=['dataFeatures', 'dataType', 'null', 'nullPct', 'duplicated', 'unique', 'uniqueSample'],
                     data=order_items_check)
order_items_desc
```
